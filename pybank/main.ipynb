```python
#Total Months: 25
#Total Revenue: $1241412
#Average Revenue Change: $216825
#Greatest Increase in Revenue: Sep-16 ($815531)
#Greatest Decrease in Revenue: Aug-12 ($-652794)

import numpy as np
import pandas as pd

file = r'budget_data_1.csv'
df = pd.read_csv(file)
print('Total Months: ', df['Date'].count())
print('Total Revenue: $', df['Revenue'].sum())
df['RevChg'] = df['Revenue'] - df['Revenue'].shift(+1)
print('Average Revenue Change: $', df['RevChg'].mean())
print('Greatest Increase in Revenue:', df['RevChg'].max())
print('Greatest Decrease in Revenue:', df['RevChg'].min())
print(df, file=open("output.txt", "a"))
df
```
